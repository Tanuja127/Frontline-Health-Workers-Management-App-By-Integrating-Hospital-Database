```python
import pandas as pd
from datetime import datetime,timedelta,time
import numpy as np
df=pd.read_csv("D:\eyantra\Book2.csv")
```

#### ARRANGING IN ASCENDING ORDER ACCORDING TO MEDICATION TIME

```python
df['medicine']=pd.to_datetime(df['medicine'],format='%H:%M:%S')
df.sort_values(by = 'medicine', ascending = True,inplace=True)
df
```

```python
times=df['medicine'].tolist()
timesec=[]
for i in range(len(times)-1):
    diff=times[i+1]-times[i]
    timesec.append(diff)
timesec.append('1000000000')
df['timediff']=timesec
```

```python
listt=df['timediff'].tolist()
newlist=[]
for i in listt:
    b=i / np.timedelta64(1, 's')
    newlist.append(b)
df.drop('timediff', 1, inplace=True)
df['original sec']=newlist
a=[]
for i in newlist:
    if(i>300):
        b=i-300
    else:
        b=i
    a.append(b)
df['5minprior']=a  
df['medicine']=pd.to_datetime(df['medicine'],format='%H:%M:%S').dt.time 
```

```python
timelist=df['medicine'].tolist()
secondslist=df['5minprior'].tolist()
names=df['name'].tolist()
```

```python
timest=[]
for i in timelist:
    abcd=i.strftime('%H:%M:%S')
    timest.append(abcd)
checkduplicate=[] #list which has no redundant timestamp
for i in timest:
    if (i not in checkduplicate):
        checkduplicate.append(i)
```
